```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
# Data import and split
data = pd.read_csv("Naive-Bayes-Classification-Data.csv")
print(data)
X = data.drop('diabetes', axis=1)
y = data['diabetes']
# print(X)
# print(y)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=40)
```

```python
def evaluate_model(model_name, y_test, y_pred, y_prob=None):
    return {
        "Model": model_name,
        "Accuracy": metrics.accuracy_score(y_test, y_pred),
        "Precision": metrics.precision_score(y_test, y_pred),
        "Recall": metrics.recall_score(y_test, y_pred),
        "F1-Score": metrics.f1_score(y_test, y_pred),
        "AUC": metrics.roc_auc_score(y_test, y_prob) if y_prob is not None else np.nan
    }

results = []
```

```python
# Naive Bayes with Laplacian correction

from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()
gnb.fit(X_train, y_train)

y_pred = gnb.predict(X_test)
y_prob = gnb.predict_proba(X_test)[:, 1]

results.append(
    evaluate_model("Naive Bayes", y_test, y_pred, y_prob)
)
print(results[-1])
```

```python
# Linear Regression with Stochastic Gradient Descent

from sklearn.linear_model import SGDClassifier

sgd = SGDClassifier(loss="log_loss", penalty="l2", max_iter=1000, eta0=0.01, random_state=45)

sgd.fit(X_train, y_train)


y_pred = sgd.predict(X_test)
y_prob = sgd.predict_proba(X_test)[:, 1]

results.append(
    evaluate_model("SGD (Linear Regression)", y_test, y_pred, y_prob)
)
print(results[-1])
```

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(X_train, y_train)

y_pred = log_reg.predict(X_test)
y_prob = log_reg.predict_proba(X_test)[:, 1]

results.append(
    evaluate_model("Logistic Regression", y_test, y_pred, y_prob)
)
print(results[-1])
```

```python
from sklearn.tree import DecisionTreeClassifier

dt_entropy = DecisionTreeClassifier(criterion="entropy", random_state=42)
dt_entropy.fit(X_train, y_train)

y_pred = dt_entropy.predict(X_test)
y_prob = dt_entropy.predict_proba(X_test)[:, 1]

results.append(
    evaluate_model("Decision Tree (Entropy)", y_test, y_pred, y_prob)
)
print(results[-1])
```

```python
import pandas as pd

# Load dataset
df = pd.read_csv("spam.csv", encoding='latin-1')

# Keep only relevant columns
df = df[['v1', 'v2']]
df.columns = ['label', 'message']

df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['label'] = le.fit_transform(df['label'])

df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split

# Convert text to numerical features
vectorizer = TfidfVectorizer(
    max_features=5000,
    stop_words='english'
)

X_txt = vectorizer.fit_transform(df['message'])
y_txt = df['label']

# Train-test split
X_train_txt, X_test_txt, y_train_txt, y_test_txt = train_test_split(
    X_txt, y_txt, test_size=0.2, random_state=42
)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.regularizers import l1


model = Sequential([
    Dense(128, activation='relu'),
    Dense(1, activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.fit(
    X_train_txt.toarray(),
    y_train_txt,
    epochs=10,
    batch_size=32,
    validation_split=0.2
)
```

```python
history = model.fit(
    X_train_txt.toarray(), y_train_txt,
    epochs=10,
    batch_size=32,
    validation_split=0.2,
    verbose=1
)
```

```python
from sklearn import metrics
import numpy as np

# Predictions
y_prob_txt = model.predict(X_test_txt.toarray()).ravel()
y_pred_txt = (y_prob_txt > 0.5).astype(int)

# Metrics
accuracy = metrics.accuracy_score(y_test_txt, y_pred_txt)
precision = metrics.precision_score(y_test_txt, y_pred_txt)
recall = metrics.recall_score(y_test_txt, y_pred_txt)
f1 = metrics.f1_score(y_test_txt, y_pred_txt)
auc = metrics.roc_auc_score(y_test_txt, y_prob_txt)

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1-score:", f1)
print("AUC:", auc)

print("\nClassification Report:\n",
      metrics.classification_report(y_test_txt, y_pred_txt))
```

```python
from sklearn import metrics
import matplotlib.pyplot as plt

fpr, tpr, _ = metrics.roc_curve(y_test_txt, y_prob_txt)

plt.figure()
plt.plot(fpr, tpr, label=f"AUC = {auc:.2f}")
plt.plot([0, 1], [0, 1], linestyle='--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve – Spam Classification (Neural Network)")
plt.legend()
plt.show()
```
