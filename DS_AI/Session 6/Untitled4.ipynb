```python
# Basic Mathematical Graphs (Sine, Cosine, Exponential)

import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2*np.pi, 2*np.pi, 100)

plt.figure()
plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.cos(x), label='cos(x)')
plt.plot(x, np.exp(x/4), label='exp(x/4)')
plt.legend()
plt.title("Basic Mathematical Functions")
plt.show()
```

```python
import pandas as pd

df = pd.read_csv("Global_Superstore2.csv", encoding="latin1")
```

```python
# Line Chart (Sales Over Time)

df['Order Date'] = pd.to_datetime(df['Order Date'])
sales_by_date = df.groupby('Order Date')['Sales'].sum()

sales_by_date.plot()
plt.title("Sales Over Time")
plt.show()
```

```python
# Scatter Plot (Sales vs Profit)

plt.scatter(df['Sales'], df['Profit'])
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.show()
```

```python
# Bar Chart (Sales by Category)

df.groupby('Category')['Sales'].sum().plot(kind='bar')
plt.title("Sales by Category")
plt.show()
```

```python
# Pie Chart (Segment Distribution)

df['Segment'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Customer Segment Distribution")
plt.show()
```

```python
# Choropleth Map (Sales by Country)

import plotly.express as px

country_sales = df.groupby('Country')['Sales'].sum().reset_index()

fig = px.choropleth(country_sales,
                    locations="Country",
                    locationmode="country names",
                    color="Sales")
fig.show()
```

```python
# Bubble Chart

plt.scatter(df['Sales'], df['Profit'],
            s=df['Quantity']*10, alpha=0.5)
plt.show()
```

```python
# Doughnut Chart

plt.pie(df['Category'].value_counts(),
        labels=df['Category'].value_counts().index,
        wedgeprops=dict(width=0.4))
plt.show()
```

```python
# Count Plot (Region)

import matplotlib.pyplot as plt

fig, axs = plt.subplots(2,3, figsize=(15,10))

# Line
sales_by_date.plot(ax=axs[0,0])
axs[0,0].set_title("Sales Over Time")

# Bar
df.groupby('Category')['Sales'].sum().plot(kind='bar', ax=axs[0,1])

# Pie
df['Segment'].value_counts().plot(kind='pie', ax=axs[0,2])

# Scatter
axs[1,0].scatter(df['Sales'], df['Profit'])

# Heatmap
sns.heatmap(df[['Sales','Profit','Quantity']].corr(),
            ax=axs[1,1], annot=True)

plt.tight_layout()
plt.show()
```
